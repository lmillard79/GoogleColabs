```
url = 'https://www.deeplearningbook.org'
ext = 'html'

def listFD(url):
    page = requests.get(url).text
    print(page)
    soup = BeautifulSoup(page, 'html.parser')
    return [url + '/' + node.get('href') for node in soup.find_all('a') if node.get('href').endswith(ext)]

for file in listFD(url):
    print(file)
```
